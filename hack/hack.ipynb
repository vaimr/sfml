```python
def prepare(df_input):
    df_output = df_input.drop(columns=[
        'price', 'description', 'fields', 'id', 'images', 'location', 
#         'name',
        'subway', 'date_created', 'category'])
    
#     categories = list(set(df_output.category))
    subcategories = list(set(df_output.subcategory))
#     df_output['category'] = df_output.category.apply(lambda r: categories.index(r))      
    df_output['subcategory'] = df_output.subcategory.apply(lambda r: subcategories.index(r))
    
    df_output = pd.get_dummies(df_output, columns=[
        'can_buy',
#         'category',
        'can_promote',
        'contacts_visible',
        'delivery_available',
        'mortgage_available',
        'payment_available'
    ])
    return df_output
                               
def preproc_data(df_input):
    df_output = df_input.copy()
    
    df_output = df_output.drop('description', axis=1)
    df_output = df_output.loc[:, ~df_output.columns.str.contains('^Unnamed')]
    df_output['name'] = df_output.name.fillna('')
    df_output['name'] = df_output['name'].str.lower()
    df_output['name'] = df_output['name'].str.replace(u'[^а-яa-z -]','')
    
    return df_output        
```
